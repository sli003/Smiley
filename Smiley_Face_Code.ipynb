```
import numpy as np
import matplotlib.pyplot as plt

def bottomhalf(x, r):
  """Calculates the bottom half of a circle."""
  y = -np.sqrt(r**2 - x**2)
  return y

def tophalf(a, b):
  """Calculates the top half of a circle."""
  z = np.sqrt(b**2 - a**2)
  return z

def sideright(c, d, height, offset):
  """Calculates the right side of a shape."""
  rightside = height + np.sqrt(d**2 - (offset + c)**2)
  return rightside

def sideleft(c, d, height, offset):
  """Calculates the left side of a shape."""
  leftside = height + np.sqrt(d**2 - (offset - c)**2)
  return leftside

line = np.linspace(-10, 10, 100)

# Adjust the range of 'line' for each plot to avoid invalid values in sqrt
plt.plot(line[np.where(np.sqrt(5)**2 - line**2 >= 0)], bottomhalf(line[np.where(np.sqrt(5)**2 - line**2 >= 0)], np.sqrt(5)))  # Plot bottom half
plt.plot(line[np.where(np.sqrt(5)**2 - line**2 >= 0)], tophalf(line[np.where(np.sqrt(5)**2 - line**2 >= 0)], np.sqrt(5)))  # Plot top half
plt.plot(line[np.where(1**2 - line**2 >= 0)], bottomhalf(line[np.where(1**2 - line**2 >= 0)], 1))  # Plot bottom half with r=1
plt.plot(line[np.where(np.sqrt(.1)**2 - (1 - line)**2 >= 0)], sideleft(line[np.where(np.sqrt(.1)**2 - (1 - line)**2 >= 0)], np.sqrt(.1), 1, 1))  # Plot left side
plt.plot(line[np.where(np.sqrt(.1)**2 - (1 + line)**2 >= 0)], sideright(line[np.where(np.sqrt(.1)**2 - (1 + line)**2 >= 0)], np.sqrt(.1), 1, 1))  # Plot right side


plt.show()  # Display the plot
```
